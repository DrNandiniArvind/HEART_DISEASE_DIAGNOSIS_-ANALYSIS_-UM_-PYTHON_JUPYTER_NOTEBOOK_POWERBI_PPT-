# Data Dictionary: 

### Attribute Information:

- Age, 
- Sex (1 = male, 0 = female)
- cp: The chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic).
- trestbps: The person's resting blood pressure (mm Hg on admission to the hospital)
- chol: The person's cholesterol measurement in mg/dl
- fbs: The person's fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false)
- restecg: Resting electrocardiographic measurement (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria)
- thalach: The person's maximum heart rate achieved
- exang: Exercise induced angina (1 = yes; 0 = no)
- oldpeak: ST depression induced by exercise relative to rest
- slope: the slope of the peak exercise ST segment (Value 1: upsloping, Value 2: flat, Value 3: down sloping)
- ca: The number of major vessels (0-3)
- thal: A blood disorder called thalassemia (3 = normal; 6 = fixed defect; 7 =reversable defect)
- num: Heart disease (0 = no, 1 = yes)

### Pairwise Scatter Plots of Integer Variables in Heart Disease Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import itertools

# Create figure for scatter plots
fig = plt.figure(figsize=(15, 12))

# List of integer columns
int_columns = [i for i in df1.columns if df1[i].dtypes == "int64"]

# Limit the number of pairs to plot
max_plots = 12  # Define a maximum number of plots
combinations = list(itertools.combinations(int_columns, 2))[:max_plots]

# Create scatter plots for the selected pairs of integer columns
for i, (col1, col2) in enumerate(combinations):
    plt.subplot(4, 3, i + 1)  # Fixed subplot arrangement
    plt.title(f'Scatter plot: {col1} vs {col2}', fontsize=14)  # Title for each scatter plot
    plt.xlabel(col1)  # Label x-axis
    plt.ylabel(col2)  # Label y-axis
    sns.scatterplot(x=df1[col1], y=df1[col2])  # Create scatter plot

# Adjust layout for better spacing
fig.tight_layout()
plt.show()
```

## Comprehensive Insights on Heart Disease Risk Factors:

###  **1. Chest Pain:**

- Type 2 chest pain is a strong predictor of heart disease, accounting for 41.63% of cases, making it a critical indicator for assessing heart health.
- Most heart disease cases involve type 2 chest pain, emphasizing its importance in diagnosis and preventive care.
- Chest pain and maximum heart rate (thalach) are directly linked to heart disease risk, highlighting these factors in diagnostic evaluations.

###  **2. Age:**

- People over 55 are predominantly affected by heart disease, with a peak incidence observed among middle-aged individuals between 40 and 45 years.
- Resting blood pressure shows an upward trend starting around age 55, rising from 130 mmHg and increasing thereafter.
- Cholesterol levels also increase gradually with age, peaking around 260-300 mg/dL after age 55 and showing fluctuations in later years.



###  **3. Gender Differences:**

- Males are more affected by heart disease than females and are more likely to experience exercise-induced angina, which is linked to heart disease.
- Females tend to have slightly higher cholesterol levels than males, while males display higher levels of ST depression and fasting blood sugar.
- Males have a higher prevalence of exercise-induced angina (exang), a condition associated with a higher heart disease risk.


###  **4. Resting Blood Pressure:**

- Resting blood pressure trends upward with age, particularly in individuals aged 55 and older.
- Blood pressure values generally cluster on the lower end but stretch to the right, indicating a subset with significantly higher values.
- There is no major gender difference in resting blood pressure levels.


###  **5. Cholesterol:**

- Elevated cholesterol is a critical risk factor for heart disease, with values highly concentrated around the mean and a smaller subset showing extremely high levels.
- In people over 50, cholesterol values demonstrate intermittent rises and falls, with a gradual upward trend beginning at age 55.


###  **6. Thalassemia:**

- Individuals with thalassemia category 2 (reversible defect) are more prone to heart disease.
- Females exhibit a relatively higher incidence of thalassemia, which could be relevant in heart disease risk analysis.


###  **7. Old Peak (ST Depression):**

- Most individuals show ST depression during exercise (oldpeak), with values clustering around the mean but with some cases showing significantly higher oldpeak levels.
- There is a strong negative correlation between oldpeak and the slope of the peak exercise ST segment (slope), indicating that high ST depression levels may be less likely as slope values increase.
- Males tend to have higher oldpeak levels than females.


###  **8. Other Factors:**

- Resting ECG values of 1 are frequently associated with heart disease, though they are also observed among individuals without heart disease, showing no significant difference in prevalence.
- The slope of the peak ST segment with a value of 2 is associated with a higher likelihood of heart disease.
- Individuals with Thalassemia type 2 (reversible defect) are more likely to develop heart disease, with females being relatively more affected.
- Higher instances of heart disease are observed in individuals over 50, with a notable count in people around 50 years old, far exceeding cases in those without heart disease.

## Recommendations for Heart Disease Prevention and Management

1. **Enhanced Screening:** Incorporate chest pain type, cholesterol levels, and maximum heart rate as standard measures in heart disease risk assessments, especially for those over 50.

    
2. **Age-Targeted Interventions:** Implement age-specific heart health programs focused on individuals aged 40-55, as this age range shows peak incidences of heart disease.

    
3. **Gender-Specific Health Plans:** Develop customized prevention strategies addressing gender differences, including managing cholesterol in females and reducing fasting blood sugar in males.

    
4. **Regular Monitoring of ST Segment and Oldpeak Levels:** Track ST depression and peak exercise ST segment levels in high-risk individuals to detect heart disease early and guide treatment.

    
5. **Promote Cholesterol and Blood Pressure Management:** Advocate for regular blood pressure and cholesterol monitoring, as these factors display strong correlations with heart disease and increase with age.
